```python
#vemos que los 0 de x, y están dentro de los 0 de z, por lo que eliminando estos valores de z también nos quitaremos los 0 de x, y.
diamonds_df = diamonds_df.drop(diamonds_df[diamonds_df['z']==0].index)
print(diamonds_df.shape)
```

```python
diamonds_df = diamonds_df.reset_index(drop=True)
diamonds_df
```

```python
diamonds_df.corr()
```

```python
plt.figure(figsize=(10,10))  
p1 = sns.heatmap(diamonds_df.corr(), annot=True, linewidths=.4, cmap='YlGnBu',square=True)
```

```python
# price, carat, x, y, z están muy correlacionadas.
sns.pairplot(diamonds_df, vars = ['carat', 'price','depth', 'x', 'y', 'z'])
```
