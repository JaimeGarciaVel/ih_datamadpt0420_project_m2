```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds_df = pd.read_csv('../data/diamonds_train.csv')
diamonds_df.head(10)
```

```python
diamonds_df.info()
```

### Challenge 1: Exploratory Data Visualization Charts and Summary Statistics

```python
#hay que quitar los valores de x, y, z que sean 0.
diamonds_df.loc[diamonds_df['x']==0]
```

```python
diamonds_df.loc[diamonds_df['y']==0]
```

```python
diamonds_df.loc[diamonds_df['z']==0]
```

```python
#vemos que los 0 de x, y están dentro de los 0 de z, por lo que eliminando estos valores de z también nos quitaremos los 0 de x, y.
diamonds_df = diamonds_df.drop(diamonds_df[diamonds_df['z']==0].index)
print(diamonds_df.shape)
```

```python
diamonds_df = diamonds_df.reset_index(drop=True)
diamonds_df
```

```python
diamonds_df.corr()
```

```python
plt.figure(figsize=(10,10))  
p1 = sns.heatmap(diamonds_df.corr(), annot=True, linewidths=.4, cmap='YlGnBu',square=True)
```

```python
# price, carat, x, y, z están muy correlacionadas. A mayor carat y volumen, mayor es el precio
sns.pairplot(diamonds_df, vars = ['carat', 'price', 'x', 'y', 'z'])
```

```python
#carat vs price + cut
plt.figure(figsize=(10,10))
plot1 = sns.scatterplot(data=diamonds_df, x="carat", y="price", hue="cut", palette="muted", style="cut")
```

```python
#carat vs price + color
plt.figure(figsize=(12,12))
plot2 = sns.scatterplot(data=diamonds_df, x="carat", y="price", hue="color", palette="muted",style="color")
```

```python
#carat vs price + clarity
plt.figure(figsize=(12,12))
plot3 = sns.scatterplot(data=diamonds_df, x="carat", y="price", hue="clarity", palette="muted",style="clarity")
```

```python
diamonds_df['cut'].value_counts().plot.bar();
```
